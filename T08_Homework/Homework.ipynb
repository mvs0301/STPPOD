### Задание 2

1. Нарисовать треугольник $ A=(-2;0), B=(1;5), C=(3;2) $<br>
2. Нарисовать треугольник $A'B'C'$, в котором изменены места координат каждой вершины: $A'(x,y)=A(y,x), B'(x,y)=B(y,x), C'(x,y)=C(y,x)$

**Решение**<br>
Найдем прямую, на которой лежит отрезок $ AB $. Уравнение прямой на плоскости $y=b_0+b_1\cdot x$. Следовательно, для решения задачи требуется определить параметры $b_0$ и $b_1$. Найдем $b_1$. При увеличении $x$ на 3 единицы (от -2 до 1) $y$ увеличивается на 5 единиц (от 0 до 5), то есть на каждую единицу $x$ имеем прирост $y$ на $\frac{5}{3}$. Тогда $b_1=\frac{5}{3}$. Найдём значение $y$, соответствующее $x=0$. Если сдвинуться от точки $B$ влево на одну единицу по оси $x$, окажемся в точке с координатой $x=0$. При этом значение $y$, равное пяти в точке $B$, уменьшится на $\frac{5}{3}$ и станет равным $5-\frac{5}{3}=\frac{10}{3}$. Тогда $b_0=\frac{10}{3}$ и искомое уравнение прямой на плоскости $y=\frac{10}{3}+\frac{5}{3}\cdot{x}$.

Построим график найденной прямой в интервале $x$ от $-2$ до $1$.

```
x_AB = np.linspace(-2, 1, 4)
y_AB = x_AB * (5/3) + (10/3)
```
Далее,
```
plt.plot(x_AB, y_AB, color='blue')
```
или
```
axes.plot(x_AB,y_AB, color='blue')
```
в зависимости от выбранного метода построения графиков. В данном примере используем объектно-ориентированый метод, использующий объект *axes*.

Построение отрезков $BC$ и $AC$ аналогично отрезку $AB$. Для построения треугольника $A'B'C'$ требуется изменить параметры функции *plt.plot().*

````python
x_AB = np.linspace(-2, 1, 4)
y_AB = x_AB * (5/3) + (10/3)

x_BC = np.linspace(1, 3, 3)
y_BC = -x_BC * (3/2) + (13/2)

x_AC = np.linspace(-2, 3, 6)
y_AC = (2/5)*x_AC + (4/5)

fig=plt.figure()
axes=fig.add_axes([0,0,1,1])

axes.plot(x_AB,y_AB, color='blue')
axes.plot(x_BC,y_BC, color='blue')
axes.plot(x_AC,y_AC, color='blue')

axes.plot(y_AB,x_AB, color='red')
axes.plot(y_BC,x_BC, color='red')
axes.plot(y_AC,x_AC, color='red')

axes.set_title('Треугольники')
axes.grid()
````
