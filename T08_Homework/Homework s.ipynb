### Задание 3

Через точки $A=(-2;0)$ и $B=(1;5)$ провести прямую в интервале $x\in {[-20;20]}$

<b>Решение</b><br>
Из курса аналитической геометрии известно, что прямая, проходящая через точки с координатами $A=(x_1;y_1)$ и $B=(x_2;y_2)$, описывается уравнением

$$\frac {y-y_1}{y_2-y_1} = \frac {x-x_1}{x_2-x_1}$$

Подставим значения координат $A$ и $B$
$$
\frac{y-0}{5-0}=\frac{x-(-2)}{1-(-2)}
$$

$$
\frac{y}{5}=\frac{x+2}{3}
$$

$$
y=\frac{5}{3}\cdot{x} + \frac{10}{3}
$$

````python
x=np.arange(-20,21,1)
y=5/3*x+10/3

fig=plt.figure()
axes=fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.plot(-2,0, color='red', marker='p')
axes.plot(1,5, color='red', marker='p')
axes.grid()
axes.set_title('Прямая, проходящая через точки A и B');
````
